# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data

```python
data = pd.read_csv('in-vehicle-coupon-recommendation.csv')
data.head()
```

# Data visualization

## Entire data

```python
# use Waterfall chart to show the data
plt.figure(figsize=(10, 10))
plt.bar(data['Y'].value_counts().index, data['Y'].value_counts(), width=0.5)
```

## 1. Property 'destination' and 'Y'

```python
sns.countplot(x='destination', data=data, hue='Y')
```

## 2. Property 'passanger' and 'Y'

```python
sns.countplot(x='passanger', data=data, hue='Y')
```

## 3. Property 'weather' and 'Y'

```python
sns.countplot(x='weather', data=data, hue='Y')
```

## 4. Property 'temperature' and 'Y'

```python
sns.countplot(x='temperature', data=data, hue='Y')
```

```python
# draw Pie chart
plt.figure(figsize=(10, 10))
plt.pie(data['temperature'].value_counts(), labels=data['temperature'].value_counts().index, autopct='%1.1f%%')
plt.show()
```
