# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data

```python
data = pd.read_table('SouthGermanCredit.asc', sep=' ')
data.head()
```

# Validate data

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

# Data visualization

## Entire data

```python
targetCol = 'kredit'
```

```python
# Use pairplot to see the data distribution
sns.pairplot(data, hue=targetCol)
```

## 1. Property 'laufkont' and 'kredit'

```python
sns.countplot(x='laufkont', data=data, hue=targetCol)
```

```python
# Plot dot chart
plt.figure(figsize=(10, 10))
plt.plot(data['laufkont'].value_counts(), data['laufkont'].value_counts().index, 'o')
plt.show()
```

## 2. Property 'passanger' and 'Y'

```python
sns.countplot(x='passanger', data=data, hue=targetCol)
```

## 3. Property 'weather' and 'Y'

```python
sns.countplot(x='weather', data=data, hue=targetCol)
```

## 4. Property 'temperature' and 'Y'

```python
sns.countplot(x='temperature', data=data, hue=targetCol)
```

```python
# draw Pie chart
plt.figure(figsize=(10, 10))
plt.pie(data['temperature'].value_counts(), labels=data['temperature'].value_counts().index,
        autopct='%1.1f%%')
plt.show()
```
