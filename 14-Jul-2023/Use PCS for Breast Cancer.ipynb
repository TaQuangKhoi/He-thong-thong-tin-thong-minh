```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer

cancer_set = load_breast_cancer()

cancer_data = pd.DataFrame(cancer_set['data'], columns=cancer_set['feature_names'])

cancer_data.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(cancer_data)

scaled_cancer_data = scaler.transform(cancer_data)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca.fit(scaled_cancer_data)

pca_scale_cancer_data = pca.transform(scaled_cancer_data)
```

```python
print("Dữ liệu gốc: ", scaled_cancer_data.shape)

print("Dữ liệu sau khi giảm chiều: ", pca_scale_cancer_data.shape)
```

```python
plt.figure(figsize=(8,6))

pca_1 = pca_scale_cancer_data[:,0]

pca_2 = pca_scale_cancer_data[:,1]

plt.scatter(pca_1, pca_2, c=cancer_set['target'], cmap='plasma')
```
