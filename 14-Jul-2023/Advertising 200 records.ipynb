```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Advertising.csv')
df
```

```python
plt.scatter(df['TV'], df['sales'], color='red')
plt.title('Sales vs TV', fontsize=14)
plt.xlabel('TV', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['radio'], df['sales'], color='red')
plt.title('Sales vs Radio', fontsize=14)
plt.xlabel('Radio', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['newspaper'], df['sales'], color='red')
plt.title('Sales vs newspaper', fontsize=14)
plt.xlabel('newspaper', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.grid(True)
plt.show()
```

# Simple Linear Regression

```python
y = df['sales']
y
```

```python
X = df[['TV', 'radio', 'newspaper']]
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
print(linreg.intercept_)
```

```python
print(linreg.coef_)
```

```python
y_pred = linreg.predict(X_test)
```

```python
from sklearn import metrics
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
linreg.score(X_test, y_test)
```
